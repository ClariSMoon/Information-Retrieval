```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
import statistics
import matplotlib.pyplot as plt
```

```python
docs = pd.read_csv('../input/psgs_w100.tsv', sep='\t')
topics = pd.read_csv('../input/sampled_train_topics.csv')
qrels = pd.read_csv('../input/sampled_train_qrels.csv')
```

```python
print(docs.shape)
docs.sample(5)
```

```python
print(topics.shape)
topics.sample(5)
```

```python
print(qrels.shape)
qrels.sample(5)
```

```python
doc_num = len(docs)

doc_max = docs['text'].str.len().max()
doc_min = docs['text'].str.len().min()
doc_avg = docs['text'].str.len().mean()
```

```python
query_num = len(topics)

query_max = topics['query'].str.len().max()
query_min = topics['query'].str.len().min()
query_avg = topics['query'].str.len().mean()
```

```python
doc_med = statistics.median(docs['text'].str.len())
query_med = statistics.median(topics['query'].str.len())
```

```python
doc_summary = pd.DataFrame({
    'Documents': [int(doc_num), int(doc_min), int(doc_max), int(doc_med), int(doc_avg)],
}, index = ['Number of items', 'Min length', 'Max length', 'Median length', 'Average length'])
```

```python
query_summary = pd.DataFrame({
    'Queries': [int(query_num), int(query_min), int(query_max), int(query_med), int(query_avg)],
}, index = ['Number of items', 'Min length', 'Max length', 'Median length', 'Average length'])
```

```python
sns.boxplot(data = docs['text'].str.len())
plt.xlabel('Length of documents')
plt.ylabel('Count')
plt.title('Box plot of document length')
plt.savefig('boxplot_docs.png')
```

```python
sns.boxplot(data = topics['query'].str.len())
plt.xlabel('Length of queries')
plt.ylabel('Count')
plt.title('Box plot of document length')
plt.savefig('boxplot_queries.png')
```
